```python
import os
path = "../output"
filenamepaths = [path+'/'+x for x in os.listdir(path)]
filenamepaths
```

```python
import pandas as pd

def read_data(filepath):
    
    df = pd.read_csv(filepath)
    # print(df.columns)
    df.rename(columns={'Unnamed: 0':'yearmon'}, inplace=True)
    # print(df.columns)
    return df
```

```python



ALL_PLOTS = {}
for f in filenamepaths:
    df = read_data(f)
    ALL_PLOTS[f[10:-4]]= df
for x,y in ALL_PLOTS.items():
    print(type(x))
    print(y.shape)
    print(type(y))
type(ALL_PLOTS)
```
