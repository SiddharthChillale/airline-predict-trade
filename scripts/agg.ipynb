```python
import pandas as pd
import numpy as np
```

```python
import os
import re

allFilePaths = {}

yearList = [int(x) for x in os.listdir('../assets')]

yearList.sort()

# yearList = yearList[:4]
```

```python

for year in yearList:
    with os.scandir('../assets/'+str(year)) as it:
        for entry in it:
            if entry.name.endswith('.csv') and entry.is_file():
                filename = entry.name
                yearID = int(str(year)+str(filename[7:9]))
                path = '../assets/'+str(year)+'/'+filename
                print(path)
                allFilePaths[yearID] = path
    
```

```python
regions = ['EASTERN', 'WESTERN', 'SOUTHERN', 'NORTHERN', 'NORTH EAST']
```

```python
headers = [x for x in range(19)]
```

```python
ix_to_month = {
    '00':"Jan",
    '01':"Feb",
    '02':"Mar",
    '03':"Apr",
    '04':"May",
    '05':"Jun",
    '06':"Jul",
    '07':"Aug",
    '08':"Sep",
    '09':"Oct",
    '10':"Nov",
    '11':"Dec"
    
}
```

```python
def separate_csv(path, yearID):
    df = pd.read_csv(path, names = headers, index_col=False  )
    
    movement = df.iloc[4:4+len(regions)]
    passengers = df.iloc[11:11+len(regions)]
    freight = df.iloc[18:18+len(regions)]
    
    movement_I, movement_D = cleanup(movement, "m",yearID)
    passengers_I, passengers_D = cleanup(passengers, "p",yearID)
    freight_I, freight_D = cleanup(freight, "f",yearID)
    
    return movement_I, movement_D, passengers_I, passengers_D, freight_I, freight_D
#     print(movement.shape)
#     print(passengers.shape)
#     print(freight.shape)
```

```python
def cleanup(dataframe, name, alt_name):
    dataframe_I = dataframe.loc[:,[0,1]]
    dataframe_D = dataframe.loc[:,[0,7]]
    
    dataframe_I.set_index(0, inplace=True)
    dataframe_D.set_index(0, inplace=True)
    
    dataframe_I.rename(columns={1:alt_name}, inplace=True)
    dataframe_D.rename(columns={7:alt_name}, inplace=True)
    
#     print(dataframe.columns)
#     if name=="m":
#         print(dataframe)
        
    return dataframe_I, dataframe_D
```

../assets/2017/output-9oct2k17annex5.csv

```python
movement_I_df = pd.DataFrame(index=regions)
passengers_I_df = pd.DataFrame(index=regions)
freight_I_df = pd.DataFrame(index=regions)

movement_D_df = pd.DataFrame(index=regions)
passengers_D_df = pd.DataFrame(index=regions)
freight_D_df = pd.DataFrame(index=regions)
```

```python

for yearID, filepath in allFilePaths.items():
#     filename = filepath[15:]
# #     print(filename)
#     year = filepath[10:14]
# #     print(year)
#     month = ix_to_month[filepath[22:24]]
#     col = month+year
#     print(col)
    movement_I, movement_D, passengers_I, passengers_D, freight_I, freight_D  = separate_csv(filepath, yearID)
    
#     print(ix, "------>\n", movement)
#     movement_df.join(movement)
    movement_I_df = movement_I_df.merge(movement_I,how='outer', left_index=True, right_index=True)
    movement_D_df = movement_D_df.merge(movement_D,how='outer', left_index=True, right_index=True)
    
    passengers_I_df = passengers_I_df.merge(passengers_I,how='outer', left_index=True, right_index=True)
    passengers_D_df = passengers_D_df.merge(passengers_D,how='outer', left_index=True, right_index=True)
    
    freight_I_df = freight_I_df.merge(freight_I,how='outer', left_index=True, right_index=True)
    freight_D_df = freight_D_df.merge(freight_D,how='outer', left_index=True, right_index=True)
    
    
#     print(ix,"--\n",movement_df)
```

```python
allData = {"movement_I_df":movement_I_df,
           "movement_D_df":movement_D_df,
           "passengers_I_df":passengers_I_df,
           "passengers_D_df":passengers_D_df,
           "freight_I_df":freight_I_df,
           "freight_D_df":freight_D_df }
```

```python
for x in allData.keys():
    allData[x] = allData[x].T
    
    allData[x].sort_index(ascending=True, inplace=True)
    print("shape of ", x, "is ", allData[x].shape)
```
