```python
import os
path = "../output"
filenamepaths = [path+'/'+x for x in os.listdir(path)]
filenamepaths
```

```python
import pandas as pd

def read_data(filepath):
    
    df = pd.read_csv(filepath, thousands=',')
    # print(df.columns)
    df.rename(columns={'Unnamed: 0':'yearmon'}, inplace=True)
    # print(df.columns)
    return df
```

```python
ALL_PLOTS = {}
for f in filenamepaths:
    df = read_data(f)
    ALL_PLOTS[f[10:-4]]= df
for x,y in ALL_PLOTS.items():
    print(type(x))
    print(y.shape)
    print(type(y))
type(ALL_PLOTS)
```

```python
import matplotlib.pyplot as plt
N_PLOTS = 6
# regions = ['EASTERN', 'WESTERN','NORTHERN', 'SOUTHERN', 'NORTH EAST']
regions = ['EASTERN']

for x,y in ALL_PLOTS.items():
    df = y

    plt.figure(figsize=(20,10))
    plot_east, = plt.plot(df.index, df['EASTERN'] ,label='EASTERN')
    plot_west, = plt.plot(df.index, df['WESTERN'] ,label='WESTERN')
    plot_north, = plt.plot(df.index, df['NORTHERN'] ,label='NORTHERN')
    #     plot_north_east, = plt.plot(df.index, df['NORTH EAST'] ,label='NORTH EAST')

    plot_south, = plt.plot(df.index, df['SOUTHERN'] , label='SOUTHERN')

    plt.xticks([x for x in range(len(df['yearmon']))], labels=df['yearmon'], rotation='vertical')
    plt.title(x)

    plt.legend([plot_east, plot_west, plot_north, plot_south], ['EAST', 'WEST', 'NORTH', 'SOUTH'])
    plt.show()
```
